## Análise gráfica

```python
sns.set_palette('Accent')
sns.set_style('dark')
```

```python
def config(ax, title, xlabel, ylabel):
    ax.figure.set_size_inches(12, 6)
    ax.set_title(title, fontsize = 18)
    ax.set_xlabel(xlabel, fontsize = 14)
    ax.set_ylabel(ylabel, fontsize = 14)
```

```python
ax = sns.boxplot(data = dados['Area'], orient = 'h', width=0.5)

config(ax, 'Preço dos Imóveis', 'Reais', '')
```

```python
ax = sns.distplot(dados['Valor'])
config(ax, 'Distribuição de Frequências', 'Valor', 'Densidade')
```

```python
ax = sns.pairplot(data=dados, y_vars=['Valor'], x_vars=['Area', 'Dist_Praia', 'Dist_Farmacia'], height=5, kind = 'reg')
```
