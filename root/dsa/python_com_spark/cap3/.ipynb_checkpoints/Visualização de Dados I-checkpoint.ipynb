# 1 - Matplotlib: Regressão Linear

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
x = np.arange(100)
y = x + np.random.choice(np.arange(-10, 10), 100)
```

```python
plt.scatter(x, y)
```

```python
slope, intercept, rvalue, pvalue, stderr = stats.linregress(x, y)
```

```python
plt.plot(x, y, 'bo')
plt.plot(x, x * slope + intercept, 'r')

plt.title('Preço da casa em função do tamanho')
plt.xlabel('Área (m²)')
plt.ylabel('Preço (R$)')
```
